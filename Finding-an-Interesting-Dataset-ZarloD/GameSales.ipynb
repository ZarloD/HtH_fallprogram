```python
import pandas

#pandas.options.display.max_columns = None
#pandas.options.display.max_rows = None

gameSales_data = pandas.read_csv("GameSales.csv")
gameSales_data
```

```python
'''
What is the highest selling game of all time?
'''

game_highest_selling = gameSales_data[['Name', 'Sales']]

game_highest_selling
```

```python
'''
What is the all time highest selling series on PC?

- separate by series making sure to exclude nan values
- add total sales by series
- sort to find highest sales by series
'''

# separate by series and sales
# add total sales
series_highest_selling = gameSales_data.groupby('Series')['Sales'].sum().sort_values(ascending=False)

series_highest_selling
```

# To find the game series with the highest sales, I separated the data by Series and Sales, added total sales of each series, and sorted by descending order. This shows that in this dataset, Minecraft is the highest selling series on PC.

```python
'''
What is the all time highest selling genre on PC?
'''

genre_highest_selling = gameSales_data[['Genre', 'Sales']]

genre_highest_selling
```

```python
'''
What year saw the highest PC game sales?
'''

release_year_highest_selling = gameSales_data.groupby('Release Year')['Sales'].sum().sort_values(ascending = False)

#test = gameSales_data.sort_values(by=['Release Year'])
#test
release_year_highest_selling
```

```python
'''
What month do PC games sell the most?
'''

release_month_highest_selling = gameSales_data.groupby('Release Month')['Sales'].sum().sort_values(ascending = False)

release_month_highest_selling
```
